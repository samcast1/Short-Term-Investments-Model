```python
def clean_review_df(city):
  df = pd.read_csv(f'{city}_reviews.csv')

  rating_df = df['rating_info'].str.split('\n', expand=True)
  rating_df.drop(columns = [1,2,4,5], inplace = True)
  rating_df.columns = ['rating','date','note']
  rating_df['rating'] = rating_df.rating.str.strip('Rating ,').str.strip(' stars')
  try:
    rating_df['rating'] = rating_df['rating'].astype('int')
  except Exception as e:
    print(f'Error converting rating column to integer for {city}: {e}')

  rating_df['date'] = ['June 2024' if 'ago' in x or 'Today' in x else x for x in rating_df['date']]
  try:
    rating_df['date'] = pd.to_datetime(rating_df['date'], format = 'mixed')
  except Exception as e:
    print(f'Error converting date column to datetime for {city}: {e}')

  df = pd.concat([rating_df, df.drop(columns = ['name','rating_info'])], axis = 1)
  df.set_index(df.date, inplace = True)
  df.drop(columns = 'date', inplace = True)
  df.reset_index(inplace = True)

  df.to_csv(f'cleaned_5/{city}_reviews_clean.csv', index = False)
  print(f'Saved {city} reviews to cleaned_5 folder.')
```

Perfect - the clean_reviews_df function works as intended. Now I need a list of every city to iterate through, and I'll vectorize the operation to speed it up.

The list of cities will come from the original zillow value indexes that I used to generate the scraping list.

```python
dfz = pd.read_csv('zhdf_top.csv')
```

```python
dfz
```

```python
cities = [x.lower().replace(', ', '-').replace(" ", "") for x in dfz.iloc[:,2:].columns]
```

```python
cities
```

```python
import concurrent.futures

def parallel_clean(cities):
    results = []

    with concurrent.futures.ProcessPoolExecutor() as executor:
      try:
        futures = [executor.submit(clean_review_df, city) for city in cities]
      except Exception as e:
        print(f"An error occurred: {e}")

      for future in concurrent.futures.as_completed(futures):
        try:
          result = future.result()
          results.append(result)
        except Exception as e:
          print(f"An error occurred: {e}")

    return results
```

```python
cleaned_files = parallel_clean(cities)
```

```python
cleaned_files[:10]
```

```python
for city in cities:
  try:
    clean_review_df(city)
  except Exception as e:
    print(f'Error cleaning {city}: {e}')
```
